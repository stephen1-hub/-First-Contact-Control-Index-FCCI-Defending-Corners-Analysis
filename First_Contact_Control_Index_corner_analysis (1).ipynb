# Define First Contact Logic

```python
key_events = [
    'Clearance',
    'Duel',
    'Shot',
    'Block',
    'Interception',
    'Ball Recovery'
]

def get_first_contact(df, idx):
    for i in range(idx+1, idx+15):
        event = df.iloc[i]

        if event['type.name'] in key_events:
            return event['type.name'], event['team.name']

    return None, None
```

```python
def get_first_contact_score(df, idx):

    for i in range(idx+1, idx+20):

        event = df.iloc[i]

        etype = event['type.name']

        if etype == 'Clearance':
            return "Clearance", 1

        elif etype == 'Interception':
            return "Interception", 1

        elif etype == 'Block':
            return "Block", 0.8

        elif etype == 'Duel':
            return "Duel", 1

        elif etype == 'Shot':
            return "Shot", -1

        elif etype == 'Goal':
            return "Goal", -2

    return "None", 0
```

# Who controls the first contact after the corner is delivered?
First Contact Control Index

# STEP 1 — Find corners

```python
corners = df[df['type.name'] == 'Pass']
corners = corners[corners['pass.type.name'] == 'Corner']
```

```python
priority = {
    "Shot": 3,
    "Clearance": 2,
    "Block": 2,
    "Interception": 2,
    "Duel": 1,
    "Ball Receipt*": 1,
    "Miscontrol": 1
}
```

```python
def get_first_contact(df, idx):

    best_event = None
    best_score = 0

    for i in range(idx+1, idx+10):

        event = df.iloc[i]
        etype = event['type.name']

        if etype in priority:

            if priority[etype] > best_score:
                best_event = (etype, event['team.name'])
                best_score = priority[etype]

    return best_event if best_event else (None, None)
```

```python
results = []

for idx in corners.index:

    event, team = get_first_contact(df, idx)

    results.append({
        "corner_index": idx,
        "event": event,
        "team": team
    })

fc_df = pd.DataFrame(results)
```

```python
def classify(event):

    if event in ["Clearance", "Block", "Interception"]:
        return "Win"

    elif event in ["Duel", "Ball Receipt*", "Miscontrol"]:
        return "Contested"

    elif event == "Shot":
        return "Danger"

    else:
        return "Other"
```

```python
fc_df["outcome"] = fc_df["event"].apply(classify)
```

# Load Multiple Match Event Files

```python
import pandas as pd
import requests
```

```python
def load_events(match_id):

    url = f"https://raw.githubusercontent.com/statsbomb/open-data/master/data/events/{match_id}.json"

    data = requests.get(url).json()

    return pd.json_normalize(data)
```

```python
priority = {
    "Shot": 3,
    "Clearance": 2,
    "Block": 2,
    "Interception": 2,
    "Duel": 1,
    "Ball Receipt*": 1,
    "Miscontrol": 1
}

def get_first_contact(df, idx):

    best_event = None
    best_score = 0

    max_index = min(idx + 10, len(df))

    for i in range(idx + 1, max_index):

        event = df.iloc[i]
        etype = event.get("type.name", None)
        team = event.get("team.name", None)

        if etype in priority and team is not None:

            if priority[etype] > best_score:
                best_event = (etype, team)
                best_score = priority[etype]

    return best_event if best_event else (None, None)
```

```python
def classify(event):

    if event in ["Clearance", "Block", "Interception"]:
        return "Win"

    elif event == "Shot":
        return "Danger"

    elif event in ["Duel", "Ball Receipt*", "Miscontrol"]:
        return "Contested"

    return "Other"
```

```python
match_ids = [3764440, 3764661, 3773369, 3773372, 3773377]

results = []

for match_id in match_ids:

    df = load_events(match_id)

    corners = df[
        (df["type.name"] == "Pass") &
        (df["pass.type.name"] == "Corner")
    ]

    for idx in corners.index:

        event, team = get_first_contact(df, idx)

        if event is None or team is None:
            continue

        results.append({
            "match_id": match_id,
            "team": team,
            "event": event,
            "outcome": classify(event)
        })

fc_df = pd.DataFrame(results)
```

```python
print(fc_df.head())
print(fc_df.columns)
print(fc_df["team"].value_counts())
```

```python
comparison = fc_df.groupby("team").agg(
    Total_Corners=("outcome", "count"),
    Wins=("outcome", lambda x: (x == "Win").sum()),
    Danger=("outcome", lambda x: (x == "Danger").sum())
).reset_index()

comparison["Win Rate"] = (comparison["Wins"] / comparison["Total_Corners"]) * 100
comparison["Danger Rate"] = (comparison["Danger"] / comparison["Total_Corners"]) * 100
comparison["FCCI"] = comparison["Win Rate"] - comparison["Danger Rate"]
```

```python
matches_url = "https://raw.githubusercontent.com/statsbomb/open-data/master/data/matches/11/90.json"
matches = pd.read_json(matches_url)

match_ids = matches["match_id"].tolist()
```

```python
results = []

for match_id in match_ids:

    df = load_events(match_id)

    corners = df[
        (df["type.name"] == "Pass") &
        (df["pass.type.name"] == "Corner")
    ]

    for idx in corners.index:

        event, team = get_first_contact(df, idx)

        if event is None or team is None:
            continue

        results.append({
            "match_id": match_id,
            "team": team,
            "event": event,
            "outcome": classify(event)
        })
```

```python
league_df = pd.DataFrame(results)

league_table = league_df.groupby("team").agg(
    Total_Corners=("outcome", "count"),
    Wins=("outcome", lambda x: (x == "Win").sum()),
    Danger=("outcome", lambda x: (x == "Danger").sum())
).reset_index()

league_table["Win Rate"] = league_table["Wins"] / league_table["Total_Corners"] * 100
league_table["Danger Rate"] = league_table["Danger"] / league_table["Total_Corners"] * 100
league_table["FCCI"] = league_table["Win Rate"] - league_table["Danger Rate"]
```
