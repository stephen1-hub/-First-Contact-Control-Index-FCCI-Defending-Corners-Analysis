# Define First Contact Logic

```python
key_events = [
    'Clearance',
    'Duel',
    'Shot',
    'Block',
    'Interception',
    'Ball Recovery'
]

def get_first_contact(df, idx):
    for i in range(idx+1, idx+15):
        event = df.iloc[i]

        if event['type.name'] in key_events:
            return event['type.name'], event['team.name']

    return None, None
```

```python
results = []
```

```python
for idx in corners.index:

    event_type, team = get_first_contact(df, idx)

    results.append({
        "corner_index": idx,
        "defending_team": team,
        "first_contact_event": event_type
    })
```

```python
def get_first_contact_score(df, idx):

    for i in range(idx+1, idx+20):

        event = df.iloc[i]

        etype = event['type.name']

        if etype == 'Clearance':
            return "Clearance", 1

        elif etype == 'Interception':
            return "Interception", 1

        elif etype == 'Block':
            return "Block", 0.8

        elif etype == 'Duel':
            return "Duel", 1

        elif etype == 'Shot':
            return "Shot", -1

        elif etype == 'Goal':
            return "Goal", -2

    return "None", 0
```

# Who controls the first contact after the corner is delivered?
First Contact Control Index

# STEP 1 — Find corners

```python
corners = df[df['type.name'] == 'Pass']
corners = corners[corners['pass.type.name'] == 'Corner']
```

```python
priority = {
    "Shot": 3,
    "Clearance": 2,
    "Block": 2,
    "Interception": 2,
    "Duel": 1,
    "Ball Receipt*": 1,
    "Miscontrol": 1
}
```

```python
def get_first_contact(df, idx):

    best_event = None
    best_score = 0

    for i in range(idx+1, idx+10):

        event = df.iloc[i]
        etype = event['type.name']

        if etype in priority:

            if priority[etype] > best_score:
                best_event = (etype, event['team.name'])
                best_score = priority[etype]

    return best_event if best_event else (None, None)
```

```python
results = []

for idx in corners.index:

    event, team = get_first_contact(df, idx)

    results.append({
        "corner_index": idx,
        "event": event,
        "team": team
    })

fc_df = pd.DataFrame(results)
```

```python
def classify(event):

    if event in ["Clearance", "Block", "Interception"]:
        return "Win"

    elif event in ["Duel", "Ball Receipt*", "Miscontrol"]:
        return "Contested"

    elif event == "Shot":
        return "Danger"

    else:
        return "Other"
```

```python
fc_df["outcome"] = fc_df["event"].apply(classify)
```

```python
win_rate = (fc_df["outcome"] == "Win").mean() * 100
```

# Process EACH match

```python
def get_first_contact(df, idx):

    priority = {
        "Shot": 3,
        "Clearance": 2,
        "Block": 2,
        "Interception": 2,
        "Duel": 1,
        "Ball Receipt*": 1,
        "Miscontrol": 1
    }

    best_event = None
    best_score = 0

    max_index = min(idx + 10, len(df))

    for i in range(idx + 1, max_index):

        event = df.iloc[i]
        etype = event['type.name']

        if etype in priority:

            if priority[etype] > best_score:
                best_event = (etype, event['team.name'])
                best_score = priority[etype]

    return best_event if best_event else (None, None)
```

```python
team_df = pd.DataFrame(all_results)
```

```python
totals = team_df.groupby("team").size()
```

```python
wins = team_df[team_df["outcome"] == "Win"].groupby("team").size()
```

```python
danger = team_df[team_df["outcome"] == "Danger"].groupby("team").size()
```

```python
comparison = pd.DataFrame({
    "Total Corners": totals,
    "Wins": wins,
    "Danger": danger
}).fillna(0)
```

```python
comparison["Win Rate"] = (
    comparison["Wins"] /
    comparison["Total Corners"]
) * 100
```

```python
comparison["Danger Rate"] = (
    comparison["Danger"] /
    comparison["Total Corners"]
) * 100
```

```python
comparison["FCCI"] = (
    comparison["Win Rate"] -
    comparison["Danger Rate"]
)
```

```python
comparison = comparison.sort_values(
    by="FCCI",
    ascending=False
)
```
